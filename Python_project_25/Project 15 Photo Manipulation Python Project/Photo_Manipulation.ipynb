```python
# 📦 Step 1: Install Pillow
!pip install Pillow

# 📤 Step 2: Upload Image
from google.colab import files
uploaded = files.upload()

# 🖼️ Step 3: Load and Display Image
from PIL import Image, ImageEnhance, ImageFilter
import matplotlib.pyplot as plt

image_name = next(iter(uploaded))
img = Image.open(image_name)

plt.figure(figsize=(5, 5))
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
plt.show()

# 🎨 Step 4: Apply Filters - Contrast, Brightness, Blur
# Increase Contrast
contrast = ImageEnhance.Contrast(img)
contrast_img = contrast.enhance(1.8)

# Increase Brightness
brightness = ImageEnhance.Brightness(contrast_img)
bright_img = brightness.enhance(1.5)

# Apply Gaussian Blur
blur_img = bright_img.filter(ImageFilter.GaussianBlur(radius=2))

# Display Edited Image
plt.figure(figsize=(5, 5))
plt.imshow(blur_img)
plt.axis('off')
plt.title("Edited Image")
plt.show()

# 💾 Step 5: Save and Download Final Image
blur_img.save("edited_image.jpg")
files.download("edited_image.jpg")
```
