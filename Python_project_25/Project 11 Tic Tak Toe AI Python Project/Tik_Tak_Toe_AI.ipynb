```python
import random

# Khali board banate hain
board = [" " for _ in range(9)]

# Board ko print karne wali function
def print_board():
    print()
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--+---+--")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--+---+--")
    print(f"{board[6]} | {board[7]} | {board[8]}")
    print()

# Check karo koi jeeta hai ya nahi
def check_winner(player):
    win_combos = [
        [0,1,2], [3,4,5], [6,7,8],  # rows
        [0,3,6], [1,4,7], [2,5,8],  # columns
        [0,4,8], [2,4,6]            # diagonals
    ]
    for combo in win_combos:
        if all(board[i] == player for i in combo):
            return True
    return False

# Check karo board full hai ya nahi (draw)
def is_full():
    return " " not in board

# Player ka move
def player_move():
    while True:
        try:
            move = int(input("Apna move daalo (1-9): ")) - 1
            if board[move] == " ":
                board[move] = "X"
                break
            else:
                print("‚ùå Yeh jagah pehle se bhari hai.")
        except:
            print("‚ùå Galat input. 1 se 9 ke beech number daalo.")

# Simple AI ka move (random empty cell)
def ai_move():
    empty = [i for i in range(9) if board[i] == " "]
    move = random.choice(empty)
    board[move] = "O"

# Game chalane wali main function
def play_game():
    print("üéÆ Tic Tac Toe - Aap vs AI")
    print("Aap: X | AI: O")
    print_board()

    while True:
        player_move()
        print_board()
        if check_winner("X"):
            print("üéâ Aap jeet gaye!")
            break
        if is_full():
            print("üòê Game draw ho gaya.")
            break

        ai_move()
        print_board()
        if check_winner("O"):
            print("üíª AI jeet gaya!")
            break
        if is_full():
            print("üòê Game draw ho gaya.")
            break

# Program yahin se start hoga
play_game()
```
